## №1 Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```
import pandas as pd
```

```
ratings=pd.read_csv('ratings.csv', sep=';')
```

```
ratings.head()
```

```
movies=pd.read_csv('movies.csv', sep=';')
```

```
movies.head()
```

## №2 Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```
top1=ratings['user_id'].value_counts().head(1)
print(top1)
```

## №3 Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```
filter_user=ratings['user_id']==405
```

```
filter_user
```

```
list405=ratings.loc[filter_user]
```

```
list405
```

```
film_items=list405.item_id
```

```
film_items
```

```
ratings=ratings.merge(
    film_items,
    left_on='item_id',
    right_on='item_id',
    how='right'
)
```

```
ratings
```

## №4 Добавьте к датафрейму из задания 3 столбцы:
-По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет

-Столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```
ratings=ratings[['user_id','item_id','rating']]
```

```
print(ratings)
```

```
df=ratings.merge(
    movies,
    left_on='item_id',
    right_on='movie_id',
    how='inner'
)
```

```
df.columns
```

```
df=df[['user_id', 'movie_id', 'rating', 'movie_title', 'release_year', 'unknown', 'Action', 'Adventure', 'Animation', "Children's",
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir',
       'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western']]
```

```
df.head()
```

```
print(df)
```

```
dfcount=df['movie_id'].value_counts()
```

```
print(dfcount)
```

```
dfcount=df[['movie_id','rating']].groupby(['movie_id']).count().sort_values(by='rating')
```

```
print(dfcount)
```

```
dfrating=df[['movie_id','rating']].groupby(['movie_id']).sum().sort_values(by='rating')
```

```
print(dfrating)
```

```
dftest=df.merge(
    dfrating,
    left_on='movie_id',
    right_on='movie_id',
    how='inner'
)
```

```
print(dftest)
```

```
dftest=dftest.merge(
    dfcount,
    left_on='movie_id',
    right_on='movie_id',
    how='inner'
)
```

```
print(dftest)
```

```
dftest.rename(columns = {'rating_x' : 'rating', 'rating_y' : 'rating_sum', 'rating' : 'rating_count'}, inplace = True)
```

```
dftest.head()
```

## №5 Сформируйте X_train, X_test, y_train, y_test

```
dftest.columns
```

```
X, y = dftest[['release_year','unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy',
       'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
       'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'rating_count',
       'rating_sum']],dftest['rating']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## №6 Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```
from sklearn.linear_model import LinearRegression
```

```
model=LinearRegression()
```

```
model.fit(X_train, y_train)
```

## №7 Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```
from sklearn.metrics import mean_squared_error
```

```
y_predict = model.predict(X_test)
mean_squared_error(y_test, y_predict)
```
